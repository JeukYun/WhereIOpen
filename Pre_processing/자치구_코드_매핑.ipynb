## 매출 데이터 불러오기

```python
df_sell = pd.read_excel('C:/Data_Engineer_Study/멀캠_PJT_1/Data_set/Raw_Data/상권 매출데이터/[탐색적분석용]21년도_34분기_매출 데이터.xlsx')
```

```python
df_sell.head()
```

```python
# df_sell['상권_코드'] = df_sell.상권_코드.map(lambda x:dic[x])
# -> 매핑하려고 했는데 코드번호 2120068번이 누락되었음
```

```python
df.loc[df['코드'] == 2120068] #2120068 누락 확인
```

```python
# 2120068번 코드 추가작업
newcol = {'자치구': '중랑구',
          '동' : '면목3.8동',
          '코드' : 2120068}
```

```python
df = df.append(newcol, ignore_index = True)
```

```python
df.to_csv('구,동별_코드_완료.csv') # 저장
```

```python
dic = dict(zip(df.코드, df.자치구))
```

```python
df_sell.insert(5,'자치구', df_sell.상권_코드.map(lambda x:dic[x]))
```

```python
df_sell.head()
```

```python
# df_sell 에 '동' 컬럼 추가
dic2 = dict(zip(df.코드, df.동))
df_sell.insert(6,'동', df_sell.상권_코드.map(lambda x:dic2[x]))
```

```python
df_sell.head()
```

```python
df_sell.to_csv('21년도_34분기_매출 데이터(자치구, 동추가).csv')
```
