# 시각화 데이터 추출

- 시간대 3구간으로 축소, 총 매출액 추가

```python
sell_all['00~11시 매출'] = sell_all['시간대_00~06_매출_금액'] + sell_all['시간대_06~11_매출_금액']
sell_all['11~17시 매출'] = sell_all['시간대_11~14_매출_금액'] + sell_all['시간대_14~17_매출_금액']
sell_all['17~24시 매출'] = sell_all['시간대_17~21_매출_금액'] + sell_all['시간대_21~24_매출_금액']
sell_all['총 매출액'] = sell_all['주중_매출_금액'] + sell_all['주말_매출_금액']
```

- 추출 할 컬럼

```python
columns = ['기준_년_코드', '자치구', '동', '상권_코드_명', '서비스_업종_코드_명', '점포수', '월요일_매출_금액', '화요일_매출_금액', 
           '수요일_매출_금액', '목요일_매출_금액', '금요일_매출_금액', '토요일_매출_금액', '일요일_매출_금액',
           '총 매출액', '주중_매출_비율', '주말_매출_비율','00~11시 매출', '11~17시 매출', '17~24시 매출']
```

```python
df = sell_all[columns]
```

```python
df.head(2)
```

- 서비스 업종이 '제과점'인 것만 추출

```python
dfB = df[df['서비스_업종_코드_명'] == '제과점']
```

## 연도별, 자치구 별 그룹화(평균)

```python
df2 = dfB.groupby(['기준_년_코드','자치구']).agg('mean')
```

```python
df2.head()
```

```python
df2.reset_index(inplace = True)
```

```python
df2.info()
```

```python
df19 = df2[df2['기준_년_코드'] == 2019]
df20 = df2[df2['기준_년_코드'] == 2020]
df21 = df2[df2['기준_년_코드'] == 2021]
df22 = df2[df2['기준_년_코드'] == 2022]
```

```python
df20 = df20.drop(columns = '기준_년_코드')
```

```python
df20.head(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 한글폰트 설정
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')
```

## Pie 차트.

- 파이차트의 경우 범주의 갯수가 적은 것이 더 눈에 띌 듯 함
- ex) 주중, 주말 매출 비율 

```python
plt.pie(df20day['강남구'], labels=df20day['강남구'].index, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
# 7개의 요일이 표기되며, 모두 비슷한 수치를 가지고 있음
# 어느 요일 모두 한눈에 들어오기 힘든 부분이 있음
```

## HeatMap -> 요일별 구의 매출 확인

```python
# 요일을 x축, 구를 y축에 배열하여 요일별 매출이 높은 구를 시각적으로 확인
df20day.T
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df20day.T, cmap='YlGnBu', annot=True, fmt=',.0f')
plt.title('서울 구별 요일별 매출 평균 히트맵')
plt.xlabel('요일')
plt.ylabel('자치구')
plt.show()
```
