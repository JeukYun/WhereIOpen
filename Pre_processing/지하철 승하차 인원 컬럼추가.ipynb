# 데이터 불러오기

```python
import pandas as pd

sub = pd.read_csv('23년 서울 지하철 하차인원 정보.csv', encoding = 'cp949')
```

```python
sub.head(3)
```

```python
sub.info()
```

```python
sub['수송일자'] = pd.to_datetime(sub['수송일자'])
```

```python
sub.info()
```

```python
sub['년'] = sub['수송일자'].dt.year
sub['월'] = sub['수송일자'].dt.month
sub['일'] = sub['수송일자'].dt.day
sub['분기'] = sub['수송일자'].dt.quarter
sub['요일'] = sub['수송일자'].dt.dayofweek
```

```python
sub
```

```python
week_list = ['월', '화', '수', '목', '금', '토', '일']

sub['요일'] = sub.apply(lambda x : week_list[x['요일']], axis = 1)
```

```python
sub
```

```python
sub.to_csv('23년 서울 지하철 하차인원 정보_컬럼추가,필터링완료.csv', encoding = 'cp949', index = False)
```
