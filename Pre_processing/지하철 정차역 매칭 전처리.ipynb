```python
# ~~역이 포함된 행 추출
raw_code = raw_code[raw_code['상권_코드_명'].str.contains('역')] 
```

## 'OO역' 뒤 필요없는 부분 제거

```python
raw_code.tail(10)
```

```python
# 공백을 기준으로 나누어도 크게 정제될거같진 않음
raw_code['상권_코드_명'].str.split(' ')
```

```python
raw_code['상권_코드_명'] = raw_code['상권_코드_명'].str.replace('역.*', '역', regex = True)

# replace('역.*', '역', regex=True) : '둔촌역전통시장'인 경우 '역'~끝까지의 단어 '역전통시장'을 '역'으로 변경, 정규표현식(.*) 사용을 위해 regex = True

# 나중에 찾은건데 '역촌중앙시장'이 '역'만 남게되었음... 
```

```python
raw_code
# 강남터미널지하도상점가2구역 -> 제거해야함
```

```python
# 글자길이가 대략 7글자 넘어가면 의심스러울듯해서 확인했으나 7글자 넘어도 정상적인 역이 많음
# 괄호 ( 가 붙어있는 데이터 앞부분을 지워야 깔끔할 듯 함.
raw_code[raw_code['상권_코드_명'].str.len() >= 7]
```

## 괄호 '(' 앞의 값 제거

```python
# '(' 가 섞인 행 추출
raw_code[raw_code['상권_코드_명'].str.contains('\(')]
```

```python
raw_code['상권_코드_명'] = raw_code['상권_코드_명'].str.split('\(').str[-1]
# str.split('\(').str[-1] '('를 기준으로 나누고 그중 마지막(-1)만을 raw_code['상권_코드_명']데이터로 받는다.
```

```python
# 제거 확인
raw_code['상권_코드_명'][[434,498,510,533]]
```

```python
raw_code
```

```python
# 공백이 있는 행 확인
raw_code[raw_code['상권_코드_명'].str.contains(" ")]
```

```python
# 공백기준 문자열 분리 후 str[-1]로 마지막 부분'독산역'선택
raw_code['상권_코드_명'] = raw_code['상권_코드_명'].str.split().str[-1]
```

```python
# 중복행 제거
raw_code.drop_duplicates(inplace=True)
```
